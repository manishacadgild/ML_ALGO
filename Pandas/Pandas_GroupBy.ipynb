```python
def run_score(batsman_name):
    new_data=data[data['batsman']==batsman_name]
    return new_data.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3).index[0]
```

```python
run_score('V Kohli')
```

```python
run_score('DA Warner')
```
