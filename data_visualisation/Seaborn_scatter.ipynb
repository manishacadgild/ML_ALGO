```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
titanic=sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
titanic.shape
```

```python
sns.scatterplot(x="age",y="fare",data=titanic)
```

```python
sns.scatterplot(x="age",y="fare" ,data=titanic)
```

```python
sns.scatterplot(x="age",y="fare" ,data=titanic,hue='sex')
```

```python
sns.scatterplot(x="age",y="fare" ,data=titanic,hue='sex',style='who')
```

```python
sns.scatterplot(x="age",y="fare" ,data=titanic,hue='sex',style='who',size='who')
```

```python
sns.scatterplot(x="age",y="fare" ,data=titanic,hue='sex',style='who',size='who',sizes=(100,200))
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x="age",y="fare" ,data=titanic,hue='sex',style='who',size='who',sizes=(100,200))
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x="age",y="fare" ,data=titanic,hue='alive',style='who',size='who',sizes=(100,200))
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x="who",y="fare" ,data=titanic,hue='alive',style='alive',size='who',sizes=(100,200))
```
