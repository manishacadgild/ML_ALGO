```python
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Pie plot

```python
area=['marketing','budget','Sales','R&D']
budget=[12,40,50,60]
plt.pie(budget,labels=area,autopct="%0.2f%%")
```

```python
x=[2,4,56,78,80,50]
y=[2,5,6,7,8,98]
plt.scatter(x,y)
plt.xlabel("X axis")
plt.ylabel("y axis")
```

```python
from matplotlib import style
style.use("ggplot")
product_id=[1,2,3,4,5,6]
online_price=[200,3000,4000,50000,6000,400]
offline_price=[300,400,600,700,900,1000]

plt.plot(product_id,online_price,'ro--',linestyle='dashed',linewidth=2,markersize=15,label='online_price')
plt.plot(product_id,offline_price,'go--',linestyle='--',linewidth=2,markersize=10,label='offline_price')
plt.xlabel("Product id")
plt.ylabel("Product price")
plt.title("Product distribution of the price")
plt.legend(loc=4)
plt.grid(color='r', linestyle='-', linewidth=2)
plt.show()
```

#### Histogram

```python
import numpy as np
import random
```

```python
ml_studentAge=np.random.randint(18,45,(100))
py_studentAge=np.random.randint(18,40,(100))
```

```python
print(ml_studentAge)
print(py_studentAge)
```

```python
bins=[15,20,30,40,45]
plt.hist(ml_studentAge,bins,rwidth=0.8,orientation ='vertical',color='blue',label='ml_studentAge')
plt.hist(py_studentAge,bins,rwidth=0.8,orientation ='horizontal',color='yellow',label='py_studentAge')
plt.title("Student ml")

plt.xlabel("Student age categories")
plt.ylabel("no of student age")
plt.legend()
```

#### Bar chart

```python
class1=['Python','Java','Kafka','ds']
student_class1=[80,50,56,70]
```

```python
plt.bar(class1,student_class1)
```

```python
plt.barh(class1,student_class1)
```

```python
plt.bar(class1,student_class1,width=0.2,align='edge',color="y",edgecolor='m',alpha=0.8,linewidth=5,linestyle='--'
        ,label="Class_Student")
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.bar(class1,student_class1,width=0.6,align='edge',color="k",edgecolor='m',alpha=0.8,linewidth=5,linestyle='--'
        ,label="Class_Student")
```
