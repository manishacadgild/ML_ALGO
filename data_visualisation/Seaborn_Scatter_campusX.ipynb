```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
plt.style.use('fivethirtyeight')
```

```python
data=sns.load_dataset('tips')
```

```python
data.head()
```

#### plot a scatter plot between total_bill and tips

```python
### plot here
sns.relplot(x='total_bill',y='tip',kind='scatter',data=data)
```

```python
### use hue parameter used for categorical data
sns.relplot(x='total_bill',y='tip',hue='smoker',kind='scatter',data=data)
```

```python
#### Style
sns.relplot(x='total_bill',y='tip',hue='smoker',kind='scatter',style='sex',data=data)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',kind='scatter',style='sex',size='size',sizes=(20,40),data=data)
```

```python
sns.catplot(x='day',y='total_bill',hue='smoker',kind='box',data=data)
```
