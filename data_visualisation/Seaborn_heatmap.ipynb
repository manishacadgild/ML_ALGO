```python
group=pd.read_csv("Who_is_responsible_for_global_warming.csv")
group.head()
```

```python
group  = group.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code'], axis=1).set_index('Country Name')
group
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(group)
```

```python
sns.heatmap(group)
```

```python
plt.figure(figsize=(16,10))
sns.heatmap(group,vmin=0,vmax=21)
```

```python
plt.figure(figsize=(16,10))
sns.heatmap(group,vmin=0,vmax=21,cmap="GnBu",annot=True)
```
